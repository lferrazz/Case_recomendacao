```python
#Importanto bibliotecas para fazer a modelagem
from sklearn.feature_extraction.text import TfidfVectorizer
#Cossine_similarity, biblioteca que consegue ver e comparar a similiaridade dos produtos
from sklearn.metrics.pairwise import cosine_similarity
```

```python
#Transformando a coluna "about_product" em vetores,
tfidf = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf.fit_transform(df['about_product'].fillna(''))
```

```python
#Agora criar o calcular a similiaridade dos produtos usando o cosine_similarity
cosine_similarity = cosine_similarity(tfidf_matrix, tfidf_matrix)
```

```python
indices = pd.Series(df.index, index=df['product_name']).drop_duplicates()

def recommend_product(product_name, num_recommendations=5):
    if product_name not in indices:
        return f"Produto '{product_name}' não encontrado."

    # Índice do produto
    idx = indices[product_name].iloc[0] if isinstance(indices[product_name], pd.Series) else indices[product_name]

    # Calcular a similaridade do produto com todos os outros
    sim_scores = list(enumerate(cosine_similarity[idx]))

    # Ordenar pela similaridade (descendente)
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

    # Pegar os top produtos similares (ignorando o próprio)
    product_indices = [i[0] for i in sim_scores[1:num_recommendations+1]]

    # Retornar os produtos recomendados ordenados por rating
    return df.loc[product_indices, ['product_name', 'rating']].sort_values(by='rating', ascending=False)
```
